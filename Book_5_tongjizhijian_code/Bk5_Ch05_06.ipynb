该代码计算并展示了几何分布的统计特征（均值、方差、偏度、峰度）和其概率质量函数（PMF）与累积分布函数（CDF）在不同参数下的变化情况。首先设定成功概率 \( p = 0.5 \)，计算并输出几何分布的统计量，然后绘制 PMF 和 CDF，分别展示成功概率 \( p \) 固定时 PMF 和 CDF 的变化。接着，代码绘制了不同 \( p \) 值下的几何分布 PMF，以观察成功概率变化对分布形状的影响。

几何分布的概率质量函数 (PMF) 和累积分布函数 (CDF) 分别为：
$$
P(X = k) = (1 - p)^{k - 1} \cdot p
$$
$$
P(X \leq k) = 1 - (1 - p)^k
$$

```python
from scipy.stats import geom
import matplotlib.pyplot as plt
import numpy as np
```

```python
p = plt.rcParams  # 获取默认绘图参数
p["font.sans-serif"] = ["Roboto"]  # 设置无衬线字体为 Roboto
p["font.weight"] = "light"  # 设置字体粗细为 light
p["ytick.minor.visible"] = True  # 显示 y 轴的小刻度
p["xtick.minor.visible"] = True  # 显示 x 轴的小刻度
p["axes.grid"] = True  # 启用坐标轴网格
p["grid.color"] = "0.5"  # 设置网格颜色为灰色
p["grid.linewidth"] = 0.5  # 设置网格线宽度
```

## 设置k的取值范围

```python
k_range = np.arange(1, 21)  # 设置k的取值范围
```

## 可视化

```python
fig, ax = plt.subplots(figsize = (8,6))  # 创建绘图窗口

p_values = np.arange(0.1, 1.0, 0.1)  # 定义不同的p值
num_colors = len(p_values)  # 计算颜色数量

for idx, p in enumerate(p_values):  # 遍历不同的p值
    color = plt.cm.hsv(idx / num_colors)  # 根据HSV色谱生成颜色
    plt.plot(
        k_range,
        geom.pmf(k_range, p),  # 计算并绘制PMF曲线
        marker='x',
        markersize=8,
        label='p = {:.1f}'.format(p),  # 添加标签标识p值
        color=color,  # 设置线条颜色
    )

plt.xlabel('x')  # 设置x轴标签
plt.ylabel('PMF')  # 设置y轴标签
# plt.gca().spines['right'].set_visible(False)  # 隐藏右边框
# plt.gca().spines['top'].set_visible(False)  # 隐藏上边框
# plt.gca().yaxis.set_ticks_position('left')  # y轴刻度在左侧
# plt.gca().xaxis.set_ticks_position('bottom')  # x轴刻度在底部
plt.xlim(1, k_range.max())  # 设置x轴范围
plt.ylim(0, 0.8)  # 设置y轴范围
plt.xticks([1, 5, 10, 15, 20])  # 设置x轴刻度
plt.legend()  # 显示图例
plt.show()  # 显示图形
```
