# 色子点数之和为6
Book_5《统计至简》 | 鸢尾花书：从加减乘除到机器学习 (第二版)

该代码模拟了投掷两个骰子若干次，并统计出两骰子点数之和为6的概率随着投掷次数增加的收敛趋势。主要过程如下：

1. 首先，代码生成了两个大小为 \( \text{num\_toss} = 10,000,000 \) 的数组，分别表示每次投掷两个骰子的结果。对于每个骰子，其值范围为1到6，即：
   $$ x = \text{np.random.randint}(1, 7, \text{size} = \text{num\_toss}) $$
   $$ y = \text{np.random.randint}(1, 7, \text{size} = \text{num\_toss}) $$

2. 然后计算两个骰子的点数和为6的次数，并记录每次投掷之后和为6的累计出现次数。通过累积求和实现这一操作：
   $$ \text{sum\_6} = \text{np.cumsum}((x + y) == 6) $$
   其中 \( (x + y) == 6 \) 是一个布尔数组，当且仅当两个骰子的和为6时返回True，对应于1，否则为0。

3. 根据累积结果，计算每次投掷后的和为6的概率。这个概率等于和为6的累计出现次数除以总投掷次数：
   $$ \text{prob\_sum\_6} = \frac{\text{sum\_6}}{\text{num\_toss\_array}} $$
   其中 \( \text{num\_toss\_array} \) 是从1到总投掷次数的数组，用于将每次投掷的结果和累计次数匹配。

4. 最后，通过 `matplotlib` 和 `plotly` 库进行可视化。使用对数坐标系展示概率随投掷次数增加的收敛趋势，使我们可以观察到和为6的概率逐渐逼近理论概率 \( \frac{5}{36} \approx 0.1389 \)：

   - 使用 `matplotlib` 绘制简单的概率收敛曲线；
   - 使用 `plotly` 绘制同样的曲线并添加标题和轴标签，进一步优化显示效果。

```python
import numpy as np
import matplotlib.pyplot as plt 
```

## 设置投掷次数

```python
num_toss = 10000000 # 总投掷次数设为10000000
x = np.random.randint(1, 7, size = num_toss) # 随机生成第一个骰子的投掷结果
y = np.random.randint(1, 7, size = num_toss) # 随机生成第二个骰子的投掷结果
```

```python
num_toss_array = np.arange(1, num_toss + 1) # 生成投掷次数的数组
```

```python
sum_6 = np.cumsum((x + y) == 6); # 计算和为6的累计次数
```

```python
prob_sum_6 = sum_6 / num_toss_array; # 计算每次投掷后和为6的概率
```

## 可视化

```python
fig, ax = plt.subplots()
plt.plot(num_toss_array, prob_sum_6) # 绘制投掷次数与概率的关系曲线

ax.set_xscale('log') # 设置x轴为对数坐标
plt.xlabel('Number of tosses'); # 设置x轴标签
plt.ylabel('Probability') # 设置y轴标签
ax.grid(linestyle='--', linewidth=0.25, color=[0.5,0.5,0.5]) # 设置网格线样式
```
