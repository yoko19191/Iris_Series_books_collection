**代码解释：**
此代码展示了不同尺度参数 $\sigma$ 下的对数正态分布的概率密度函数（PDF）和累积分布函数（CDF）。对数正态分布的 PDF 和 CDF 分别定义如下：

- **概率密度函数（PDF）**：

  $$
  f(x; \mu, \sigma) = \frac{1}{x \sigma \sqrt{2\pi}} \exp\left(-\frac{(\ln(x) - \mu)^2}{2\sigma^2}\right), \quad x > 0
  $$

  其中，$\mu$ 为位置参数，$\sigma$ 为尺度参数。

- **累积分布函数（CDF）**：

  $$
  F(x; \mu, \sigma) = \frac{1}{2} \left[1 + \operatorname{erf} \left(\frac{\ln(x) - \mu}{\sigma \sqrt{2}}\right)\right]
  $$

### 代码流程：

1. **绘制 PDF 曲线**：展示不同 $\sigma$ 值下对数正态分布的 PDF 曲线，显示随着 $\sigma$ 的增大，分布的扩展性增强。
   
2. **绘制 CDF 曲线**：展示不同 $\sigma$ 值下对数正态分布的 CDF 曲线，较大的 $\sigma$ 值使得 CDF 曲线变得更加平缓，表明分布的累积速度变慢。

```python
import matplotlib.pyplot as plt 
import numpy as np
from scipy.stats import lognorm
from matplotlib import cm # Colormaps
```

```python
x = np.linspace(start=0, stop=10, num=500)  # 定义 x 轴范围
```

## 绘制 PDF 曲线

```python
fig, ax = plt.subplots()  # 创建 PDF 图像

STDs = np.arange(0.5, 2.1, 0.1)  # 不同的尺度参数 σ 值
colors = plt.cm.RdYlBu_r(np.linspace(0, 1, len(STDs)))  # 使用配色方案

for i in range(len(STDs)):
    std = STDs[i]  # 当前尺度参数 σ
    plt.plot(x, lognorm.pdf(x, loc=0, s=std),  # 绘制对数正态分布的 PDF
             color=colors[int(i)],
             label="\u03C3= %.1f" % std)

plt.ylim((0, 1.5))  # 设置 y 轴范围
plt.xlim((0, 10))  # 设置 x 轴范围
plt.title("PDF of lognormal")  # 图标题
plt.ylabel("PDF")  # y 轴标签
plt.legend()  # 显示图例
plt.show()  # 显示图像
```

## 绘制 CDF 曲线

```python
fig, ax = plt.subplots()  # 创建 CDF 图像

for i in range(len(STDs)):
    std = STDs[i]  # 当前尺度参数 σ
    plt.plot(x, lognorm.cdf(x, loc=0, s=std),  # 绘制对数正态分布的 CDF
             color=colors[int(i)],
             label="\u03C3= %.1f" % std)

plt.ylim((0, 1))  # 设置 y 轴范围
plt.xlim((0, 10))  # 设置 x 轴范围
plt.title("CDF of lognormal")  # 图标题
plt.ylabel("CDF")  # y 轴标签
plt.legend()  # 显示图例
plt.show()  # 显示图像
```
