# Seaborn中的二元KDE

这段代码使用 Seaborn 和 Scikit-learn 的鸢尾花数据集，直观地展示了萼片长度 $X_1$ 和萼片宽度 $X_2$ 的联合分布，目的是观察不同品种 (setosa, versicolor, virginica) 的分布特征。

1. **数据加载与准备**：代码首先从 Seaborn 内置的数据集中载入鸢尾花数据集 `iris_sns`，并选择特征“萼片长度” (`sepal_length`) 和“萼片宽度” (`sepal_width`) 作为坐标轴，用于可视化不同品种的数据分布。

2. **散点图与核密度估计**：
   - 第一个 `jointplot` 使用散点图，展示不同品种的样本在 $X_1$ 和 $X_2$ 特征上的分布。不同品种使用不同的颜色展示，具体颜色设定为 setosa (`#FF3300`), versicolor (`#0099FF`), virginica (`#8A8A8A`)，以方便区分各个品种的聚集特征。
   - 第二个 `jointplot` 使用联合核密度估计 (KDE) 图来估计每个品种在 $X_1$ 和 $X_2$ 空间的概率密度。KDE 显示的是各个品种的联合概率密度函数 $f_{X_1, X_2}(x_1, x_2)$，表示为：
     $$
     f_{X_1, X_2}(x_1, x_2) = \frac{1}{nh^2} \sum_{i=1}^n K\left(\frac{x_1 - X_{1i}}{h}\right) K\left(\frac{x_2 - X_{2i}}{h}\right)
     $$
     其中 $K$ 为核函数，$h$ 为带宽。不同颜色表示各品种的联合密度分布，帮助直观理解各品种特征分布在特征空间中的集中程度和离散度。

通过散点图与 KDE 图的结合，代码帮助分析萼片长度和宽度特征对品种的影响，以及不同品种之间的分布差异。

```python
import numpy as np
import statsmodels.api as sm
import matplotlib.pyplot as plt 
import pandas as pd  
from sklearn.datasets import load_iris
import scipy.stats as st
import seaborn as sns  # 导入seaborn库
```

## 使用seaborn可视化数据

```python
# 加载iris数据集
iris_sns = sns.load_dataset("iris")  # 从Seaborn加载iris数据集
```

```python
# 绘制花萼长度和宽度的散点图
sns.jointplot(data=iris_sns, x="sepal_length", y="sepal_width", hue='species',
              palette={'setosa': '#FF3300', 'versicolor': '#0099FF', 'virginica': '#8A8A8A'})
```

```python
# 绘制花萼长度和宽度的核密度估计图
sns.jointplot(data=iris_sns, x="sepal_length", y="sepal_width", hue='species',
              kind="kde",  # 使用核密度估计图
              palette={'setosa': '#FF3300', 'versicolor': '#0099FF', 'virginica': '#8A8A8A'})
```
