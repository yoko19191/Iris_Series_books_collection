该代码计算并可视化了不同试验次数 \(K\) 下的二项分布概率质量函数 (PMF)，并在图中标出每次试验的期望值。给定每次独立试验成功的概率 \(p = 0.7\)，通过设置 \(K\) 的不同值来观察二项分布形状随试验次数增加的变化。对于每个 \(K\)，代码计算二项分布的PMF和期望值，并在PMF图上标记期望值的垂直线。

期望值公式为：
$$ E(X) = \sum_{x=0}^{K} x \cdot p_X(x) $$



```python
from scipy.stats import binom
import matplotlib.pyplot as plt
import numpy as np
```

## 设置不同的试验次数数组

```python
KK = [1, 2, 3, 4, 8, 16, 32, 64] # 定义不同的K值，表示试验次数
p = 0.7 # 设置每次试验成功的概率为0.7
```

## 遍历不同的K值，计算并绘制对应的PMF

```python
for K in KK:
    
    x = np.arange(0, K + 1) # 生成从0到K的取值范围
    
    p_x = binom.pmf(x, K, p) # 计算PMF，即每个x值对应的概率
    
    E_x = np.sum(p_x * x) # 计算期望值
    
    fig, ax = plt.subplots() # 创建绘图窗口
    plt.stem(x, p_x) # 绘制PMF的茎叶图
    plt.axvline(x=E_x, color='r', linestyle='--') # 在期望值处添加红色虚线
    
    plt.xticks(np.arange(K + 1)) # 设置x轴刻度
    plt.xlabel('X = x') # 设置x轴标签
    plt.ylabel('PMF, $p_X(x)$') # 设置y轴标签
    plt.ylim([0, p + 0.05]) # 设置y轴范围
```
