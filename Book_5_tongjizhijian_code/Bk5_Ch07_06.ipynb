# Beta分布

这段代码用于生成 Beta 分布在不同参数组合下的概率密度函数（PDF）和累积分布函数（CDF）图像。Beta 分布是一种定义在区间 $[0, 1]$ 上的连续概率分布，其概率密度函数（PDF）为：

$$
f(x; \alpha, \beta) = \frac{x^{\alpha - 1}(1 - x)^{\beta - 1}}{B(\alpha, \beta)}
$$

其中 $\alpha > 0$ 和 $\beta > 0$ 是形状参数，$B(\alpha, \beta)$ 是 Beta 函数，定义为：

$$
B(\alpha, \beta) = \int_0^1 t^{\alpha - 1} (1 - t)^{\beta - 1} \, dt
$$

该代码在 $\alpha$ 和 $\beta$ 取值为 $\{0.1, 0.5, 1, 2, 4\}$ 的不同组合下，分别绘制 Beta 分布的 PDF 和 CDF 图像。累积分布函数（CDF）定义为：

$$
F(x; \alpha, \beta) = \int_0^x f(t; \alpha, \beta) \, dt
$$

代码通过使用网格创建多个子图，每个子图对应一个不同的 $(\alpha, \beta)$ 组合，并绘制相应的 PDF 和 CDF 曲线。通过图像可以观察到不同的 $\alpha$ 和 $\beta$ 值如何影响 Beta 分布的形状。

```python
from scipy.stats import beta
import matplotlib.pyplot as plt
import numpy as np
```

```python
x_array = np.linspace(0,1,200) # 定义 x 的范围
alpha_array = [0.1, 0.5, 1, 2, 4] # 定义不同的 alpha 值
beta_array = [0.1, 0.5, 1, 2, 4] # 定义不同的 beta 值
alpha_array_, beta_array_ = np.meshgrid(alpha_array, beta_array) # 创建 alpha 和 beta 的网格矩阵
```

## PDF曲线

```python
fig, axs = plt.subplots(nrows=5, ncols=5, figsize=(20, 20)) # 创建5x5的子图

for alpha_idx, beta_idx, ax in zip(alpha_array_.ravel(), beta_array_.ravel(), axs.ravel()):
    title_idx = '\u03B1 = ' + str(alpha_idx) + '; \u03B2 = ' + str(beta_idx) # 设置标题显示 alpha 和 beta 值
    ax.plot(x_array, beta.pdf(x_array, alpha_idx, beta_idx), 'b', lw=1) # 绘制 PDF 曲线
    ax.set_title(title_idx) # 设置子图标题
    ax.set_xlim(0,1) # 设置 x 轴范围
    ax.set_ylim(0,4) # 设置 y 轴范围
    ax.set_xticks([0,0.5,1]) # 设置 x 轴刻度
    ax.set_yticks([0,2,4]) # 设置 y 轴刻度
    ax.spines.right.set_visible(False) # 隐藏右侧边框
    ax.spines.top.set_visible(False) # 隐藏顶部边框
    ax.yaxis.set_ticks_position('left') # y 轴刻度在左侧
    ax.xaxis.set_ticks_position('bottom') # x 轴刻度在底部
    ax.tick_params(axis="x", direction='in') # x 轴刻度向内
    ax.tick_params(axis="y", direction='in') # y 轴刻度向内
```

## CDF曲线

```python
fig, axs = plt.subplots(nrows=5, ncols=5, figsize=(20, 20)) # 创建5x5的子图

for alpha_idx, beta_idx, ax in zip(alpha_array_.ravel(), beta_array_.ravel(), axs.ravel()):
    title_idx = '\u03B1 = ' + str(alpha_idx) + '; \u03B2 = ' + str(beta_idx) # 设置标题显示 alpha 和 beta 值
    ax.plot(x_array, beta.cdf(x_array, alpha_idx, beta_idx), 'b', lw=1) # 绘制 CDF 曲线
    ax.set_title(title_idx) # 设置子图标题
    ax.set_xlim(0,1) # 设置 x 轴范围
    ax.set_ylim(0,1) # 设置 y 轴范围
    ax.set_xticks([0,0.5,1]) # 设置 x 轴刻度
    ax.set_yticks([0,0.5,1]) # 设置 y 轴刻度
    ax.spines.right.set_visible(False) # 隐藏右侧边框
    ax.spines.top.set_visible(False) # 隐藏顶部边框
    ax.yaxis.set_ticks_position('left') # y 轴刻度在左侧
    ax.xaxis.set_ticks_position('bottom') # x 轴刻度在底部
    ax.tick_params(axis="x", direction='in') # x 轴刻度向内
    ax.tick_params(axis="y", direction='in') # y 轴刻度向内
```
