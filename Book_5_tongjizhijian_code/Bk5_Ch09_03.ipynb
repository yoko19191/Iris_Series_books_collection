这段代码使用标准正态分布（均值 $\mu = 0$，标准差 $\sigma = 1$）来绘制概率密度函数（PDF）和累积分布函数（CDF），并通过不同的点和分位数来展示其变化。

1. **标准正态分布的 PDF 和等距点**：
   - 首先，代码计算标准正态分布的 PDF 表达式：
     $$
     f_Z(z) = \frac{1}{\sqrt{2\pi} \sigma} e^{-\frac{(z - \mu)^2}{2\sigma^2}}
     $$
   - 在指定的点 $z$ 上，PDF 值通过 `norm.pdf(x)` 计算。通过在 $z$ 轴上均匀选择的点，绘制这些点对应的 PDF 值，从而展示了正态分布的对称性。

2. **标准正态分布的 CDF 和等距点**：
   - 接下来，代码计算 CDF 表达式：
     $$
     F_Z(z) = \int_{-\infty}^{z} f_Z(t) \, dt
     $$
   - 对应等距的 $z$ 值，CDF 通过 `norm.cdf(x)` 计算，这些点的分布展示了标准正态分布 CDF 随 $z$ 变化的平滑递增趋势。

3. **标准正态分布的 PDF 和等距分位数**：
   - 代码进一步选择了一组等距的百分位数，例如 $0.025$ 到 $0.975$。对应每个百分位数，利用 `norm.ppf(percentile)` 反推出 $z$ 值，从而获得分位数下的 PDF 值。通过这种方式展示标准正态分布下不同百分位数的密度。

4. **标准正态分布的 CDF 和等距分位数**：
   - 最后一步展示了 CDF 和等距分位数的对应关系。每个分位数对应的 CDF 值通过水平方向的线段表示分位数的概率累积。

总体而言，这段代码通过 PDF 和 CDF 展示了正态分布下 $z$ 值随 $\mu$ 和 $\sigma$ 变化的影响，并通过等距点与等距百分位数展示了分布的对称性与累积概率的变化。

这段代码展示了标准正态分布的概率密度函数 (PDF) 和累积分布函数 (CDF) 的变化。首先通过均匀选取的 $x$ 值在 PDF 和 CDF 上标记点，然后通过不同百分位数的 $x$ 值展示在 PDF 和 CDF 上的变化。这些可视化用于展示不同位置上 PDF 和 CDF 的值。

### 标准正态分布的 PDF 和 CDF 公式
1. **概率密度函数 (PDF)**:
   $$
   f_{Z}(z) = \frac{1}{\sqrt{2 \pi}} \exp \left( -\frac{z^2}{2} \right)
   $$
2. **累积分布函数 (CDF)**:
   $$
   F_{Z}(z) = \frac{1}{2} \left[ 1 + \text{erf} \left( \frac{z}{\sqrt{2}} \right) \right]
   $$

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import cm  # 导入颜色映射
from scipy.stats import norm
```

```python
X = np.arange(-3, 3, 0.05)  # 创建 X 的线性空间
mu = 0  # 设置均值
sigma = 1  # 设置标准差
```

## 标准正态分布的 PDF 和均匀分布的 x 值

```python
x_selected = np.linspace(-2.8, 2.8, num=29)  # 创建均匀选取的 x 值

f_x = norm.pdf(X, loc=mu, scale=sigma)  # 计算标准正态分布的 PDF
```

```python
fig, ax = plt.subplots()

plt.plot(X, f_x)  # 绘制 PDF 曲线

colors = plt.cm.RdYlBu_r(np.linspace(0, 1, len(x_selected)))  # 设置颜色渐变

for i in np.linspace(0, len(x_selected) - 1, len(x_selected)):  # 遍历选取的 x 值
    
    x_selected_i = x_selected[int(i)]  # 当前 x 值
    x_PDF = norm.pdf(x_selected_i)  # 计算当前 x 值的 PDF 值

    plt.vlines(x=x_selected_i, ymin=0, ymax=x_PDF, color=colors[int(i)])  # 绘制垂直线
    plt.plot(x_selected_i, x_PDF, marker='x', color=colors[int(i)], markersize=12)  # 标记点
    
ax.set_xlim(-3, 3)  # 设置 X 轴范围
ax.set_ylim(0, 0.5)  # 设置 Y 轴范围
ax.set_xlabel('z')  # 设置 X 轴标签
ax.set_ylabel(r'PDF, $f_{Z}(z)$')  # 设置 Y 轴标签
```

## 标准正态分布的 CDF 和均匀分布的 x 值

```python
F_x = norm.cdf(X, loc=mu, scale=sigma)  # 计算标准正态分布的 CDF
```

```python
fig, ax = plt.subplots()

plt.plot(X, F_x)  # 绘制 CDF 曲线

colors = plt.cm.RdYlBu_r(np.linspace(0, 1, len(x_selected)))  # 设置颜色渐变

for i in np.linspace(0, len(x_selected) - 1, len(x_selected)):  # 遍历选取的 x 值
    
    x_selected_i = x_selected[int(i)]  # 当前 x 值
    x_CDF = norm.cdf(x_selected_i)  # 计算当前 x 值的 CDF 值

    plt.vlines(x=x_selected_i, ymin=0, ymax=x_CDF, color=colors[int(i)])  # 绘制垂直线
    plt.hlines(y=x_CDF, xmin=-3, xmax=x_selected_i, color=colors[int(i)])  # 绘制水平线
    plt.plot(x_selected_i, x_CDF, marker='x', color=colors[int(i)], markersize=12)  # 标记点
    
ax.set_xlim(-3, 3)  # 设置 X 轴范围
ax.set_ylim(0, 1)  # 设置 Y 轴范围
ax.set_xlabel('X')  # 设置 X 轴标签
ax.set_ylabel(r'PDF, $f_{Z}(z)$')  # 设置 Y 轴标签
```

## 标准正态分布的 PDF 和均匀选取的百分位数

```python
percentiles = np.linspace(0.025, 0.975, num=39)  # 创建均匀分布的百分位数
```

```python
f_x = norm.pdf(X, loc=mu, scale=sigma)  # 计算标准正态分布的 PDF
```

```python
fig, ax = plt.subplots()

plt.plot(X, f_x)  # 绘制 PDF 曲线

colors = plt.cm.RdYlBu_r(np.linspace(0, 1, len(percentiles)))  # 设置颜色渐变

for i in np.linspace(0, len(percentiles) - 1, len(percentiles)):  # 遍历百分位数
    
    percentile = percentiles[int(i)]  # 当前百分位数
    x_percent = norm.ppf(percentile)  # 计算当前百分位数的 x 值
    x_PDF = norm.pdf(x_percent, loc=mu, scale=sigma)  # 计算当前百分位数的 PDF 值
    
    plt.vlines(x=x_percent, ymin=0, ymax=x_PDF, color=colors[int(i)])  # 绘制垂直线
    plt.plot(x_percent, x_PDF, marker='x', color=colors[int(i)], markersize=12)  # 标记点
    
ax.set_xlim(-3, 3)  # 设置 X 轴范围
ax.set_ylim(0, 0.45)  # 设置 Y 轴范围
ax.set_xlabel('z')  # 设置 X 轴标签
ax.set_ylabel(r'PDF, $f_{X}(z)$')  # 设置 Y 轴标签
```

## 标准正态分布的 CDF 和百分位数

F_x = norm.cdf(X, loc=mu, scale=sigma)  # 计算标准正态分布的 CDF

```python
fig, ax = plt.subplots(figsize = (6,8))

plt.plot(F_x, X)  # 绘制 CDF 曲线

colors = plt.cm.RdYlBu_r(np.linspace(0, 1, len(percentiles)))  # 设置颜色渐变

for i in np.linspace(0, len(percentiles) - 1, len(percentiles)):  # 遍历百分位数
    
    percentile = percentiles[int(i)]  # 当前百分位数
    x_percent = norm.ppf(percentile)  # 计算当前百分位数的 x 值
    plt.vlines(x=percentile, ymin=-3, ymax=x_percent, color=colors[int(i)])  # 绘制水平线
    plt.hlines(y=x_percent, xmin=0, xmax=percentile, color=colors[int(i)])  # 绘制垂直线
    plt.plot(percentile, x_percent, marker='x', color=colors[int(i)], markersize=12)  # 标记点

ax.set_ylim(-3, 3)  # 设置 X 轴范围
ax.set_xlim(0, 1)  # 设置 Y 轴范围
ax.set_ylabel('z')  # 设置 X 轴标签
ax.set_xlabel(r'CDF, $F_{Z}(z)$')  # 设置 Y 轴标签
```
