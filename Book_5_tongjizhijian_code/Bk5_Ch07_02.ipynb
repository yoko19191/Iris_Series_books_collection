这段代码展示了不同尺度参数 $s$ 下逻辑分布的概率密度函数（PDF）和累积分布函数（CDF）。通过改变尺度参数 $s$，观察逻辑分布形状的变化。逻辑分布的 PDF 和 CDF 分别定义如下：

- **概率密度函数（PDF）**：
  
  $$
  f(x; \mu, s) = \frac{e^{-\frac{x - \mu}{s}}}{s \left(1 + e^{-\frac{x - \mu}{s}}\right)^2}
  $$
  
  其中 $\mu$ 为位置参数，$s$ 为尺度参数。

- **累积分布函数（CDF）**：

  $$
  F(x; \mu, s) = \frac{1}{1 + e^{-\frac{x - \mu}{s}}}
  $$

### 代码执行流程：

1. **绘制 PDF 曲线**：在同一图中展示不同 $s$ 值（从 0.5 到 2.0 之间）对应的逻辑分布 PDF 曲线。可以观察到，随着 $s$ 值增大，PDF 曲线变得更加平缓，表示分布的扩展性增强。
   
2. **绘制 CDF 曲线**：在同一图中展示不同 $s$ 值对应的逻辑分布 CDF 曲线。较大的 $s$ 值使得 CDF 曲线变得更加平滑，表明分布的累积速度变慢。

```python
import matplotlib.pyplot as plt 
import numpy as np
from scipy.stats import logistic
from matplotlib import cm # Colormaps
```

```python
x = np.linspace(start = -5, stop = 5, num = 200)  # 定义 x 轴范围
```

## 绘制 PDF 曲线

```python
fig, ax = plt.subplots()  # 创建 PDF 图像

Ss = np.arange(0.5, 2.1, 0.1)  # 定义 scale 参数 s 的取值范围
colors = plt.cm.RdYlBu_r(np.linspace(0, 1, len(Ss)))  # 使用配色方案

for i in range(0, len(Ss)):
    s = Ss[i]  # 当前的 scale 值
    plt.plot(x, logistic.pdf(x, loc=0, scale=s),  # 绘制不同 s 值的 PDF
             color=colors[int(i)],
             label="s = %.1f" % s)

ax.axvline(x=0, color='k', linestyle='--')  # 添加均值线

plt.ylim((0, 0.5))  # 设置 y 轴范围
plt.xlim((-5, 5))  # 设置 x 轴范围
plt.title("PDF of logistic distribution")  # 图标题
plt.ylabel("PDF")  # y 轴标签
plt.legend()  # 显示图例
plt.show()  # 显示图像
```

## 绘制 CDF 曲线

```python
fig, ax = plt.subplots()  # 创建 CDF 图像

for i in range(0, len(Ss)):
    s = Ss[i]  # 当前的 scale 值
    plt.plot(x, logistic.cdf(x, loc=0, scale=s),  # 绘制不同 s 值的 CDF
             color=colors[int(i)],
             label="s = %.1f" % s)

ax.axvline(x=0, color='k', linestyle='--')  # 添加均值线
plt.ylim((0, 1))  # 设置 y 轴范围
plt.xlim((-5, 5))  # 设置 x 轴范围
plt.title("CDF of logistic distribution")  # 图标题
plt.ylabel("CDF")  # y 轴标签
plt.legend()  # 显示图例
plt.show()  # 显示图像
```
