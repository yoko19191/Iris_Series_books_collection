# 回归分析
Book_5《统计至简》 | 鸢尾花书：从加减乘除到机器学习 (第二版)

该代码使用 `seaborn` 的鸢尾花数据集（`iris`），创建了一系列成对特征变量的散点图和核密度估计图（KDE图），以探索不同特征之间的关系。代码先绘制了无标签的 `pairplot`，并设置回归线的颜色为红色，同时调整散点透明度，然后为每个特征对添加了线性回归线。接着，在图中添加了基于花种类的色彩编码（`hue="species"`），以便区分不同花种类的数据分布。

每个特征对的回归方程为：
$$
y = a + bx
$$
其中 \( a \) 为截距，\( b \) 为回归系数，由最小二乘法估计。对角线上的图展示了各特征的核密度估计分布，以更清晰地显示数据在不同花种类下的特征分布。通过这种成对特征的可视化分析，可以观察到花萼长度、花萼宽度、花瓣长度、花瓣宽度等特征在不同花种类下的线性关系及其分布差异。

## 导入所需的库

```python
import seaborn as sns  # 导入数据可视化库Seaborn
```

## 加载鸢尾花数据集

```python
iris_sns = sns.load_dataset("iris")  # 使用Seaborn加载自带的iris数据集
```

## 绘制变量关系图（不分种类）

```python
# 创建一个成对变量关系图，显示回归线与核密度估计
g = sns.pairplot(
    iris_sns, 
    kind='reg',  # 设置关系图中的回归类型
    diag_kind='kde',  # 设置对角线上显示核密度估计
    plot_kws={
        'line_kws': {'color': 'red'},  # 设置回归线的颜色为红色
        'scatter_kws': {'alpha': 0.5}  # 设置散点的透明度为0.5
    }
)
```

## 绘制变量关系图（按种类分类）

```python
# 创建一个按种类分类的成对变量关系图，显示回归线与核密度估计
g = sns.pairplot(
    iris_sns, 
    kind='reg',  # 设置关系图中的回归类型
    diag_kind='kde',  # 设置对角线上显示核密度估计
    hue="species",  # 按species列的种类进行颜色区分
    plot_kws={'scatter_kws': {'alpha': 0.5}}  # 设置散点的透明度为0.5
)
```
