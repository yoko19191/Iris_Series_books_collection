# 向量运算

这段代码旨在通过可视化向量的相加、相减和标量乘法操作来直观地展示二维向量运算的基本概念。首先，定义了向量 $ \vec{a} = [4, 1] $ 和 $ \vec{b} = [1, 3] $，并在二维坐标系中绘制这两个向量。

1. **向量相加**：计算向量 $ \vec{a} + \vec{b} $，其中
   $$
   \vec{a} + \vec{b} = [4 + 1, 1 + 3] = [5, 4]
   $$
   代码将原始向量 $ \vec{a} $ 和 $ \vec{b} $ 以及结果向量 $ \vec{a} + \vec{b} $ 绘制在同一图中，从而展示向量加法的效果。

2. **向量相减**：计算 $ \vec{a} - \vec{b} $，其中
   $$
   \vec{a} - \vec{b} = [4 - 1, 1 - 3] = [3, -2]
   $$
   代码在坐标系中绘制了 $ \vec{a} $、$ \vec{b} $ 和 $ \vec{a} - \vec{b} $，展示了向量相减的几何意义。

3. **标量乘法**：在标量乘法部分，使用标量 $2$ 和 $0.5$ 分别与向量 $ \vec{a} = [2, 2] $ 相乘，得到
   $$
   2 \cdot \vec{a} = [4, 4] \quad \text{和} \quad 0.5 \cdot \vec{a} = [1, 1]
   $$
   通过绘制 $ 2\vec{a} $、$ \vec{a} $ 和 $ 0.5\vec{a} $，可以观察到标量乘法如何放大或缩小向量的大小，而方向保持不变。

每部分都通过图示展示运算结果，直观地理解向量在坐标平面上的几何意义。

## 导入包

```python
import numpy as np
import matplotlib.pyplot as plt
```

## 从原点绘制向量的函数

```python
def draw_vector(vector, RBG, label):  
    array = np.array([[0, 0, vector[0], vector[1]]])  # 向量的起点与终点坐标
    X, Y, U, V = zip(*array)
    plt.quiver(X, Y, U, V, angles='xy', scale_units='xy', scale=1, color=RBG)  # 绘制向量
    
    # 添加标签显示向量坐标
    label = label + f" ({vector[0]},{vector[1]})"
    plt.annotate(label,  
                 (vector[0], vector[1]),  
                 textcoords="offset points",  
                 xytext=(0, 10),  
                 ha='center')  
```

## 定义两个向量 a 和 b

```python
a = np.array([4, 1])  # 向量 a
b = np.array([1, 3])  # 向量 b
```

## 向量相加 a + b

```python
fig, ax = plt.subplots()

draw_vector(a, np.array([0, 112, 192]) / 255, 'a')  # 绘制向量 a
draw_vector(b, np.array([255, 0, 0]) / 255, 'b')    # 绘制向量 b
draw_vector(a + b, np.array([146, 208, 80]) / 255, 'a + b')  # 绘制 a + b

plt.xlabel('$x$')  # 设置 x 轴标签
plt.ylabel('$y$')  # 设置 y 轴标签

plt.axis('scaled')
ax.set_xlim([0, 5])  # 设置 x 轴范围
ax.set_ylim([0, 5])  # 设置 y 轴范围
ax.grid(linestyle='--', linewidth=0.25, color=[0.5, 0.5, 0.5])  # 设置网格线
plt.show()
```

## 向量相减 a - b

```python
fig, ax = plt.subplots()

draw_vector(a, np.array([0, 112, 192]) / 255, 'a')   # 绘制向量 a
draw_vector(b, np.array([255, 0, 0]) / 255, 'b')     # 绘制向量 b
draw_vector(a - b, np.array([146, 208, 80]) / 255, 'a - b')  # 绘制 a - b

plt.xlabel('$x$')  # 设置 x 轴标签
plt.ylabel('$y$')  # 设置 y 轴标签

plt.axis('scaled')
ax.set_xlim([0, 5])  # 设置 x 轴范围
ax.set_ylim([-2, 3])  # 设置 y 轴范围
ax.grid(linestyle='--', linewidth=0.25, color=[0.5, 0.5, 0.5])  # 设置网格线
plt.show()
```

## 标量乘法

```python
a = np.array([2, 2])  # 定义向量 a
```

```python
fig, ax = plt.subplots()

draw_vector(2 * a, np.array([146, 208, 80]) / 255, '2*a')    # 绘制 2*a
draw_vector(a, np.array([0, 112, 192]) / 255, 'a')           # 绘制 a
draw_vector(0.5 * a, np.array([255, 0, 0]) / 255, '0.5*a')   # 绘制 0.5*a

plt.xlabel('$x$')  # 设置 x 轴标签
plt.ylabel('$y$')  # 设置 y 轴标签

plt.axis('scaled')
ax.set_xlim([0, 5])  # 设置 x 轴范围
ax.set_ylim([0, 5])  # 设置 y 轴范围
ax.grid(linestyle='--', linewidth=0.25, color=[0.5, 0.5, 0.5])  # 设置网格线
plt.show()
```
