# 三维空间两点之间距离
Book_3《数学要素》 | 鸢尾花书：从加减乘除到机器学习 (第二版)

这段代码通过Matplotlib在三维空间中绘制了两个点$A$和$B$，并连接它们形成一条线段，以展示三维空间中的点与点之间的几何关系。具体步骤如下：

1. **定义点$A$和点$B$的坐标**：点$A$的坐标为$(-2, 3, 1)$，点$B$的坐标为$(3, -1, -3)$，其中$x, y, z$分别代表每个点在三维空间中的位置。点$A$和点$B$的空间向量可以记作：
   $$
   \vec{A} = (-2, 3, 1)
   $$
   $$
   \vec{B} = (3, -1, -3)
   $$

2. **计算线段的方向向量**：两点连成的线段的方向向量由$\vec{B} - \vec{A}$给出，其公式为：
   $$
   \vec{AB} = (x_B - x_A, y_B - y_A, z_B - z_A) = (3 - (-2), -1 - 3, -3 - 1) = (5, -4, -4)
   $$
   该方向向量$\vec{AB}$描述了从点$A$到点$B$的方向。

3. **计算线段的长度**：线段$AB$的长度由向量$\vec{AB}$的模（长度）给出：
   $$
   |\vec{AB}| = \sqrt{(x_B - x_A)^2 + (y_B - y_A)^2 + (z_B - z_A)^2} = \sqrt{5^2 + (-4)^2 + (-4)^2} = \sqrt{25 + 16 + 16} = \sqrt{57}
   $$
   这个长度描述了点$A$与点$B$在三维空间中的欧几里得距离。

4. **三维可视化**：在图形中，点$A$和点$B$被绘制成标记点，并用一条线段连接它们，形成直观的空间连接。代码采用正交投影（`ortho`），保证在三维视角下没有透视变形，使得各轴保持真实比例。此外，设置了坐标轴的范围为$[-4, 4]$，并添加浅灰色网格辅助观察。

5. **视角调整**：代码将视角设置为方位角$-135$度和仰角$30$度，从这一角度可以清晰观察到点$A$、点$B$以及它们的连线在三维坐标系中的位置和结构。

通过这些步骤，图形展示了点$A$和点$B$之间的距离和空间关系，是空间几何中点与点的连接、方向向量和距离计算的可视化呈现。

## 导入所需库

```python
import numpy as np  # 导入numpy库，用于数值计算
import matplotlib.pyplot as plt  # 导入matplotlib库，用于绘图
```

## 定义三维空间中的点A和点B

```python
x_A = -2  # 点A的x坐标
y_A = 3   # 点A的y坐标
z_A = 1   # 点A的z坐标

x_B = 3   # 点B的x坐标
y_B = -1  # 点B的y坐标
z_B = -3  # 点B的z坐标
```

## 创建图形

```python
fig = plt.figure()  # 创建图形对象
ax = plt.axes(projection='3d')  # 创建三维坐标轴对象
ax.plot([x_A, x_B], [y_A, y_B], [z_A, z_B], 'x', linestyle='-')  # 绘制点A和点B，并连接两点的线段

ax.set_proj_type('ortho')  # 设置投影类型为正交投影

## 设置坐标轴范围和标签
ax.set_xlim(-4, 4)  # 设置x轴的显示范围为-4到4
ax.set_ylim(-4, 4)  # 设置y轴的显示范围为-4到4
ax.set_zlim(-4, 4)  # 设置z轴的显示范围为-4到4

plt.tight_layout()  # 自动调整布局
ax.set_xlabel(r'$\it{x}$')  # 设置x轴标签
ax.set_ylabel(r'$\it{y}$')  # 设置y轴标签
ax.set_zlabel(r'$\it{z}$')  # 设置z轴标签

## 设置视角和网格样式
ax.view_init(azim=-135, elev=30)  # 设置视角，方位角为-135度，仰角为30度
ax.xaxis._axinfo["grid"].update({"linewidth":0.25, "linestyle" : ":"})  # 设置x轴网格线样式
ax.yaxis._axinfo["grid"].update({"linewidth":0.25, "linestyle" : ":"})  # 设置y轴网格线样式
ax.zaxis._axinfo["grid"].update({"linewidth":0.25, "linestyle" : ":"})  # 设置z轴网格线样式
```
