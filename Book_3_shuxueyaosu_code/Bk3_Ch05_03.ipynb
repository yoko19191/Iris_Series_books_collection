这段代码使用Matplotlib和Numpy库绘制了两个方程的图形，找到它们的交点，并对图形做了详细的样式设置，呈现出两个变量的关系及其解。代码的主要功能如下：

1. **设置图形尺寸和布局**：首先调整了图形大小和自动布局，使内容在$7.5 \times 3.5$英寸的图形窗口中显示。自动布局避免了内容重叠，使图形更清晰。

2. **定义数据点和线性方程**：
   - 使用$np.arange(0, 30 + 1, step=1)$生成从$0$到$30$的自变量数组$x\_array$，步长为$1$。
   - 定义了两个方程来表示两条直线的关系：
     - $$y\_line\_1 = 35 - x\_array$$
     - $$y\_line\_2 = \frac{94 - 2 \cdot x\_array}{4}$$
   - 这两个方程分别表示了两条直线的关系，将它们绘制到同一个图形上，展示了它们在二维平面上的变化趋势。

3. **绘制两条直线**：在同一图形上绘制了$y\_line\_1$和$y\_line\_2$。$y\_line\_1$用蓝色绘制，$y\_line\_2$用绿色绘制。这样可以清晰地看到它们在平面坐标系中的关系。

4. **标记交点**：在图中明确标记了这两个方程的解，即交点位置$(23, 12)$。在该点上使用“x”标记，并且在$x=23$和$y=12$处分别绘制垂直和水平虚线，虚线颜色为红色。这些线条指向交点，使该点在图形中更易辨识。

5. **设置坐标轴标签**：对$x$轴和$y$轴分别添加标签$x_1$和$x_2$，用以表示方程中的两个变量。$x_1$的含义为“number of chickens”，$x_2$的含义为“number of rabbits”。

6. **添加$x$轴和$y$轴的基准线**：在$x=0$和$y=0$的位置分别绘制了水平和垂直基准线。这些基准线帮助确定方程和交点的位置，同时便于观察坐标轴的对称性。

7. **设置刻度和比例**：
   - 将$x$和$y$轴的刻度范围限制在$[0, 30]$，步长为$5$，确保图形在给定的范围内显示。
   - 调用$plt.axis('scaled')$使$x$轴和$y$轴的单位比例一致，以避免图形的拉伸或缩放失真。

8. **启用小刻度**：开启了小刻度，使得网格更加细化，辅助阅读数据。

9. **网格线设置**：图形中加入了两种不同的网格线样式：
   - 较浅的灰色虚线用于小刻度，辅助定位。
   - 较深的灰色虚线用于主要网格，增强坐标的可读性。

10. **隐藏坐标轴边框**：通过隐藏图形四周的边框，使得图形更加简洁，突出展示了交点和方程关系。

## 导入所需库

```python
import numpy as np  # 导入numpy库，用于数值计算
import matplotlib.pyplot as plt  # 导入matplotlib库，用于绘图
```

## 设置图形参数

```python
plt.rcParams["figure.figsize"] = [7.50, 3.50]  # 设置图形大小为7.5 x 3.5英寸
plt.rcParams["figure.autolayout"] = True  # 设置自动布局以避免重叠
```

## 定义数据

```python
x_array = np.arange(0, 30 + 1, step=1)  # 生成从0到30的x坐标数组，步长为1
y_line_1 = 35 - x_array  # 定义y_line_1的方程：y = 35 - x
y_line_2 = (94 - 2 * x_array) / 4  # 定义y_line_2的方程：y = (94 - 2x) / 4
```

## 创建图形

```python
fig, ax = plt.subplots()  # 创建一个图形和坐标轴

## 绘制两条直线
plt.plot(x_array, y_line_1, color='#0070C0')  # 绘制y_line_1，颜色为蓝色
plt.plot(x_array, y_line_2, color='g')  # 绘制y_line_2，颜色为绿色

## 标记线性方程的解
plt.plot(23, 12, marker='x', markersize=12)  # 在点(23,12)处标记解，标记样式为“x”
plt.axvline(x=23, color='r', linestyle='--')  # 绘制一条垂直于x=23的虚线，颜色为红色
plt.axhline(y=12, color='r', linestyle='--')  # 绘制一条水平于y=12的虚线，颜色为红色

## 设置坐标轴标签
plt.xlabel('$x_1$ (number of chickens)')  # 设置x轴标签为“x1 (number of chickens)”
plt.ylabel('$x_2$ (number of rabbits)')  # 设置y轴标签为“x2 (number of rabbits)”

## 绘制x轴和y轴的基准线
plt.axhline(y=0, color='k', linestyle='-')  # 在y=0处绘制水平线，颜色为黑色
plt.axvline(x=0, color='k', linestyle='-')  # 在x=0处绘制垂直线，颜色为黑色

## 设置刻度范围和比例
plt.xticks(np.arange(0, 30 + 1, step=5))  # 设置x轴刻度范围为0到30，步长为5
plt.yticks(np.arange(0, 30 + 1, step=5))  # 设置y轴刻度范围为0到30，步长为5
plt.axis('scaled')  # 设置坐标轴比例，使x和y轴单位长度一致

## 启用小刻度
plt.minorticks_on()  # 启用小刻度

## 添加网格线
ax.grid(which='minor', linestyle=':', linewidth='0.5', color=[0.8, 0.8, 0.8])  # 添加小刻度网格线，颜色为浅灰色
ax.set_xlim(0, 30)  # 设置x轴的显示范围为0到30
ax.set_ylim(0, 30)  # 设置y轴的显示范围为0到30

## 隐藏坐标轴边框
ax.spines['top'].set_visible(False)  # 隐藏顶部边框
ax.spines['right'].set_visible(False)  # 隐藏右侧边框
ax.spines['bottom'].set_visible(False)  # 隐藏底部边框
ax.spines['left'].set_visible(False)  # 隐藏左侧边框

## 添加主要网格线
ax.grid(linestyle='--', linewidth=0.25, color=[0.5, 0.5, 0.5])  # 添加虚线网格，颜色为灰色
```
